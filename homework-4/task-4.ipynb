```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
%matplotlib inline
```

```python
def func(x,a):
    return np.sin(x*a)

x_range = np.linspace(100, 500, 5)
a_range = np.linspace(0.01, 0.02, 25)
roots = set()

for x in x_range:
    for a in a_range:
        roots.add((fsolve(func, x, args=a)[0], a))

plt.xlabel('a', fontsize=16)
plt.ylabel('x', fontsize=16)
plt.xlim(a_range[0], a_range[-1])
plt.ylim(x_range[0],x_range[-1])
plt.scatter([a for _, a in roots], [x for x, _ in roots])
```

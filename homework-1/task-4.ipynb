```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 50)
coeffs = [1, 0.5]
marker_type = ['o', 'v']
for params in zip(coeffs, marker_type):
    plt.plot(x, np.cos(params[0]*x), marker=params[1], label='cos('+str(params[0])+'*x)')
plt.legend()
```

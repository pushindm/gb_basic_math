```python
# Task 1.2: Find length of vector
def vector_length(arr):
    return np.sqrt(sum([i**2 for i in arr]))

a = [10, 10, 10]
print(f"Vector length: {vector_length(a):.2f}")
```

<h2>Task 3<h2>

```python
# Task 3. Draw circle, ellipse and hyperbola. 
def calc_2D_curve(a, b, t, pts_num=300):
    '''a, b - coefficients of curve (numbers). t - type of curve (string)'''
    curve_type_dict = dict(zip(['c', 'e', 'h'], [1, 1, 0])) # c - circle, e - ellipse, h - hyperbola
    x = np.array([a*(1-i/pts_num) for i in range(pts_num+1)]) if curve_type_dict.get(t) else np.array([a+i/pts_num for i in range(pts_num)]) 
    y = np.sqrt(np.power(b,2)*(1-np.power(x/a,2))) if curve_type_dict.get(t) else np.sqrt(np.power(b,2)*(np.power(x/a,2)-1))
    return x, y

# main
colors = ['r', 'b', 'g']
for color, coeffs in zip(colors, [[2, 2, 'c'], [2, 4, 'e'], [2, 4, 'h']]):
    a, b, t = coeffs
    x, y = calc_2D_curve(a, b, t)
    plt.plot(x, y, color)
    plt.plot(-x, y, color)
    plt.plot(x, -y, color)
    plt.plot(-x, -y, color)

plt.xlabel('x')
plt.ylabel('y')
plt.axis("equal") # set equal number of pixels for both axis
plt.show()
```

<h3>Task 5.1<h3>

```python
# two parallel planes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
Z1 = 2*X+2*Y
Z2 = 2*X+2*Y-100
ax.plot_surface(X, Y, Z1, color='r', alpha = 0.5)
ax.plot_wireframe(X,Y,Z2, alpha = 1.0)
ax.scatter(0,0,0,'z',100, 'b')
plt.show()
```

<h3>Task 5.2<h3>

```python
# draw two-sheet hyperboloid
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
a, b, c = 2, 3, 4
X, Y = np.meshgrid(np.linspace(-a, a, 10),np.linspace(-b, b, 10))
Z = c*np.sqrt(np.power(X/a,2)+np.power(Y/b,2)+1)
ax.plot_surface(X, Y, Z, color='b',alpha=0.3)
ax.plot_surface(X, Y, -Z, color='b', alpha=0.3)

# draw cylinder
X, Z = np.meshgrid(np.linspace(-1, 1, 25),np.linspace(-10, 10, 25))
Y = np.sqrt(1-np.power(X,2))
ax.plot_surface(X, Y, Z, color='r')
ax.plot_surface(X, -Y, Z, color='r')
plt.show()
```

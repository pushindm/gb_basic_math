<h3>Task 3.1<h3>

```python
def polar_to_decart(r, phi):
    return r*np.cos(phi), r*np.sin(phi)

r, phi = 3, np.pi
x, y = polar_to_decart(r, phi)
print(f"The decart coords of (r={r:.2f}, phi={phi:.2f}): (x={x:.2f}, y={y:.2f})")
```

<h3>Task 3.2<h3>

```python
def draw_circle(r):
    pts_num = 50
    return np.linspace(0,2*np.pi,pts_num), r*np.ones(pts_num)

r = 2
phi, r = draw_circle(r)

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.plot(phi, r)
ax.set_ylim(0,3)
ax.set_yticks(np.arange(0,3,1.0))
plt.show()
```

<h3>Task 3.3<h3>

```python
def draw_segment(phi, l):
    pts_num = 50
    return phi*np.ones(pts_num), np.linspace(0,l,pts_num)

phi, l = np.pi/2, 3
angles, radii = draw_segment(phi, l)

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.plot(angles, radii)
ax.set_ylim(0,l + 1)
ax.set_yticks(np.arange(0,l+1,1.0))
plt.show()
```

<h2>Task 4<h2>

```python
from scipy.optimize import fsolve

# plot graph. It helps to find the initial approximation of function zeros
x = np.linspace(-2.5, 4.5, 200)
plt.plot(x, np.exp(x)+x*(2-np.power(x,2))-1) # four intersetions with Ox axis
plt.axhline(0, color='r')
plt.xlabel('x')
plt.ylabel('z')

# for finding initial approximation of y
def sec_eq(x):
    return x**2 - 1

# system of equations
def equations(x):
    return [x[1]-x[0]**2+1, np.exp(x[0])+x[0]*(1-x[1])-1]

roots = []
x0 = [-2, -0.2, 2.5, 4.1] # initial approximation of x[0] (x)
y0 = [sec_eq(x) for x in x0] # initial approximation of x[1] (y)
roots0 = [[x,y] for x,y in zip(x0,y0)]

for root0 in roots0:
    x = fsolve(equations, root0)
    roots.append(x)

print("Solution of task-4:")
for root in roots:
    print(f"(x={root[0]:.2f},y={root[1]:.2f})")

###
### task 4.1
###
# answer: x for which curve plotted below is above zero.
print("\nSolution of 4.1:")
print(f"Answer: (x < {roots[0][0]:.2f}, y > {roots[0][1]:.2f}) and ({roots[1][0]:.2f} < x < {roots[2][0]:.2f}, {roots[1][1]:.2f} < y < {roots[2][1]:.2f}) and (x > {roots[3][0]:.2f}, y > {roots[3][1]:.2f})")
```

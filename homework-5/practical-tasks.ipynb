```python
from math import comb, pow
```

<h3>3.1<h3>

```python
n, k = 10, 8 # for analytical estimation
sample_size, successes_number = 10000, 0 # for Monte-Carlo

random_arrs = np.array([np.random.randint(0, 2, sample_size) for i in range(n)])
x = np.sum(random_arrs, axis=0)

for i in range(0, sample_size):
    if x[i] == k:
        successes_number += 1

print(f"Analytical probability estimation: {comb(n,k) / pow(2, n):.3f}")
print(f"Numerical probability estimation: {successes_number/sample_size:.3f}")
```

<h3>3.2<h3>

```python
def calculate_probability(n, k, sample_size):
    '''Calculate probability of k success in n attempts 
       using Monter-Carlo approach with given sample_size 
       and analytical approach'''
    
    successes_number = 0
    random_arrs = np.array([np.random.randint(0, 2, sample_size) for i in range(n)])
    x = np.sum(random_arrs, axis=0)

    for i in range(0, sample_size):
        if x[i] == k:
            successes_number += 1
    
    print(f"Analytical probability estimation (n={n}, k={k}): {comb(n,k) / pow(2, n):.3f}")
    print(f"Numerical probability estimation (sample_size={sample_size}): {successes_number/sample_size:.3f}\n")

n = 10
sample_size = 5000
k = [i for i in range(0, n+1)]
for j in k:
    calculate_probability(n, j, sample_size)
```

<h2>Task 5<h2>

```python
n = 100
r = 0.8
x = np.random.rand(n)
y = r*x + (1-r)*np.random.rand(n)
plt.plot(x,y,'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
a = (np.sum(x)*np.sum(y) - n*np.sum(x*y)) / (np.sum(x)*np.sum(x) - n*np.sum(x*x))
b = (np.sum(y) - a*np.sum(x)) / n
A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y, rcond=None)[0]
print(a, b)
print(a1, b1)

# calculate regression coeff
xm = np.mean(x)
ym = np.mean(y)
R = np.sum((x - xm)*(y-ym)) / np.power(np.sum((x-xm)*(x-xm))*np.sum((y-ym)*(y-ym)),1/2)
print(f"Correlation coefficient: {R:.4f}")

plt.plot([0, 1], [b, a + b])
plt.show()
```
